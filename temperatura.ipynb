```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import datetime
import locale
locale.setlocale(locale.LC_ALL, 'italian') # used to set months names in italian 
```

```python
df = pd.read_csv("temperatura/T0129_temperatura.csv" , sep=",", index_col=0,
                skiprows=[0,1,2,3], usecols=[0,1,2], parse_dates=[0], dayfirst=True,
                names=["date","temperature", "quality_temperature"])

df = df.dropna() # remove NaN values
df = df.astype({"quality_temperature":"int64"})
df = df.loc[(df.quality_temperature == 1)] # keep only values where quality factor is 1
df.pop("quality_temperature")

df
```

```python
df.describe()
```

### Temperatura durante la notte

```python
df_notte = df.loc[(df.index.hour <= 4) | (df.index.hour >= 21)]
df_notte.groupby(df_notte.index.month).mean()
```

Mese di luglio:

```python
df_notte.loc[(df_notte.index.month==7)].describe()
```

### Temperatura durante il giorno

```python
df_giorno = df.loc[(df.index.hour >= 12) & (df.index.hour <= 15)]
df_giorno.groupby(df_giorno.index.month).mean()
```
